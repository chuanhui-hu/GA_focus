```python
from quad_tree import *
from GA_focus import *
import pygad
import matplotlib.pyplot as plt
from multiprocessing import Pool, Manager
import os
import shutil
import sys
```

```python
map_folder = "../final_test_maps/"
map_name = 'Ellipse'
```

```python
result_img = cv2.imread(f"{map_folder}{map_name}.png", cv2.IMREAD_GRAYSCALE)

START, GOAL = start_goal_dict[map_name]


for i in range(10):
    result_folder = f"../final_test_results/{map_name}/{i}"
    solution = []
    with open(f"{result_folder}/solution.txt", "r") as f:
        while True:
            line = f.readline()
            if not line:
                break
            x_str, y_str = line.split(", ")
            x = int(x_str)
            y = int(y_str)

            solution.append((x, y))
        
    result_img = draw_solution(solution, result_img, start=START, goal=GOAL)

plt.figure(figsize = (15,15))
plt.imshow(result_img)
```

```python
# plot fitness history
plt.figure()
for i in range(10):
    result_folder = f"../final_test_results/{map_name}/{i}"
    fit = []
    with open(f"{result_folder}/fitness_history.txt", "r") as f:
        while True:
            line = f.readline()
            if not line:
                break
            fit.append(float(line))
    plt.plot(fit, "b-")
```

```python
record = {}
for i in range(10):
    result_folder = f"../final_test_results/{map_name}/{i}"
    with open(f"{result_folder}/result.txt", "r") as f:
        while True:
            line = f.readline()
            if not line:
                break
            split = line.index(":")
            topic = line[:split]
            value = float(line[split+2:])
            record[topic] = record.get(topic, []) + [value]
```
